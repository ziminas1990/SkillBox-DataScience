```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Загрузка данных

```python
df = pd.read_csv('vgsales.csv')
df.head()
```

### Задача 1

Постройте линейный график динамики объёма продаж всех игр жанра «Sports» в Японии. Задайте графику размер 12 на 6.

```python
data = df[df["Genre"] == "Sports"].groupby("Year").sum()
fix, axes = plt.subplots(figsize=(12, 6))
axes.set_title("Объём продаж игр жанра Sports в Японии")
axes.set_xlabel("Год")
axes.set_ylabel("Объём продаж, млн. копий")

sns.lineplot(data=data, x="Year", y="JP_Sales", ax=axes)

plt.show()
```

### Задача 2

В одних координатных осях постройте линейные графики динамики продаж студии «Activision» в Северной Америке, Европе, Японии и во всем мире.

```python
data = df[df["Publisher"] == "Activision"].groupby("Year").sum().reset_index().melt(
    id_vars=["Year"], 
    value_vars=["NA_Sales", "JP_Sales", "EU_Sales", "Global_Sales"],
    var_name="Area",
    value_name="Sales")

fix, axes = plt.subplots(figsize=(12, 6))
axes.set_title("Объём продаж игр Activision")
axes.set_xlabel("Год")
axes.set_ylabel("Объём продаж, млн. копий")

sns.lineplot(data=data, x="Year", y="Sales", hue="Area", ax=axes)
```

### Комментарий для преподавателя:

В этом задании допустимо вместо прохода циклом использовать `pd.melt()`

### Задача 3

В цикле (т.е. на разных координатных осях) постройте линейные графики динамики продаж студии «Activision» в Северной Америке, Европе, Японии и во всем мире.

```python
areas = { 
    "NA_Sales": "North America",
    "JP_Sales": "Japan",
    "EU_Sales": "Europe",
    "Global_Sales": "Global"
}

data = df[df["Publisher"] == "Activision"].groupby("Year").sum()

fix, all_axes = plt.subplots(2, 2, figsize=(16, 12), sharex=True)

for index, area in enumerate(areas):
    axes = all_axes[index // 2][index % 2]
    sns.lineplot(data=data, x="Year", y=area, ax=axes)
    axes.set_title(f"Объём продаж в {areas[area]}")
    axes.set_ylabel(f"Объём продаж, млн. копий")
    axes.set_xlabel(f"Год")
    

plt.show()
```

### Задача 5

В одной координатной сетке постройте 2 гистограммы распредения мировых продаж игр издателей «Microsoft Game Studios» и «Take-Two Interactive» после 2010 года (включая 2010 год). Графики сделайте полупрозрачными, отсечки должны быть общими и соответствовать ширине столбцов обоих графиков.

```python
studio_filter = df["Publisher"].isin(["Microsoft Game Studios", "Take-Two Interactive"])
year_filter = df["Year"] >= 2010
data = df[studio_filter & year_filter][["Publisher", "Global_Sales"]]

fix, axes = plt.subplots(figsize=(12, 6))
axes.set_title("Гистрограмма распределение мировых продаж игр")
axes.set_xlabel("Объём продаж, млн. копий")
axes.set_ylabel("Количество игр")

sns.histplot(data=data, x="Global_Sales", hue="Publisher", ax=axes)
plt.show()
```

### Задача 6

Постройте линейный график динамики количества игр, выпускаемых Nintendo, по годам. С помощью numpy или pandas определите промежуток, в который издатель каждый год выпускал больше 35 игр, и подсветите этот промежуток времени с помощью зелёного полупрозрачного прямоугольника.

```python
# Prepare data
data = df[df["Publisher"]=="Nintendo"][["Year", "Name"]].groupby("Year").count().reset_index()
data.rename(columns={"Name": "Total"}, inplace=True)

# Prepare axes
fix, axes = plt.subplots(figsize=(12, 6))
axes.set_title("Динамика выпускаемых игр Ninetendo")
axes.set_xlabel("Год")
axes.set_ylabel("Количество игр")

# Build the plot
sns.lineplot(data=data, x="Year", y="Total", ax=axes)

#Highlight data 
highlighted_years = data[data["Total"] > 35]["Year"]
axes.axvspan(xmin=highlighted_years.min(),
             xmax=highlighted_years.max(),
             color="Green", alpha=0.3)
```

### Задача 7

Определите 3 жанра и 4 платформы с самыми большими продажами игр в мире за всё время. Постройте сетку графиков 4 на 3, в каждой ячейке постройте точечный график, проверяющий зависимость общемировых продаж от продаж в Северной Америке для каждой пары (жанра, платформы). Разными цветами укажите игры разных годов. Размер итогового полотна определите самостоятельно таким образом, чтобы все графики были видны.

Делайте графики достаточно читаемыми, чтобы из них можно было сделать вывод о наличии / отсутствии зависимостей в данных.

```python
def find_best(data, what: str, count: int, criteria: str):
    return data[[what, criteria]].groupby(what).sum().sort_values(criteria, ascending=False)\
              .head(count).reset_index()[what]

best_genres = find_best(data=df, what="Genre", count=3, criteria="Global_Sales")
best_platforms = find_best(data=df, what="Platform", count=4, criteria="Global_Sales")

columns = ["Genre", "Platform", "NA_Sales", "Global_Sales", "Year"]
data = df[(df["Genre"].isin(best_genres)) & (df["Platform"].isin(best_platforms))][columns]

grid=sns.FacetGrid(data=data, row="Platform", col="Genre", hue="Year", 
                   sharex=False, sharey=False, height=4, aspect=1)
grid.map(sns.scatterplot, "NA_Sales", "Global_Sales")
```
