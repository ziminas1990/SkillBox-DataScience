```python
from typing import Dict
import csv

# I use class just to make 'int' mutable
class Stat:
    def __init__(self):
        self.actions: int = 0
    
    def __repr__(self):
        return str(self.actions)

# Month -> Page -> Device -> Stat
Month = str
Page = str
Device = str
DevicesStat = Dict[Device, Stat]
PagesStat = Dict[Page, DevicesStat]
MonthsStat = Dict[Month, PagesStat]

statistic: MonthsStat = {}

devices = ["Mobile", "Desktop"]
    
# name.csv ниже нужно заменить на название csv файла, который вы хотите открыть
file_path = 'hw_7_click_stream2.csv'
with open(file_path, mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file, fieldnames=['ID','page','date','device'])
    for row in csv_reader:
        _, page, date, device = row.values()
        year, month, day = date.split("-")
        assert device in devices, f"Unexpected device '{device}'"
        
        month_stat = statistic.setdefault(month, {})
        page_stat = month_stat.setdefault(page, {})
        stat = page_stat.setdefault(device, Stat())
        stat.actions += 1
        
print("Month\tPage\t\tMobile\tDesktop")
months = list(statistic.keys())
months.sort()
for month in months:
    print(f"Month {month}:")
    month_stat = statistic[month]
    for page, page_stat in month_stat.items():
        print(f"\t{page}:\t{page_stat['Mobile']}\t{page_stat['Desktop']}")
```

# Задание 7.6

```python
from typing import Dict, Set, List
import csv
import math

file_path: str = 'hw_7_click_stream3.csv'
normalize: bool = True

# I use class just to make 'int' mutable
class Stat:
    def __init__(self):
        self.actions: int = 0
    
    def __repr__(self):
        return str(self.actions)

# Month -> Device -> Gender -> Page -> Stat
Month = str
Page = str
Device = str
Gender = str
PagesStat = Dict[Page, Stat]
GenderStat = Dict[Gender, PagesStat]
DevicesStat = Dict[Device, GenderStat]
MonthsStat = Dict[Month, DevicesStat]

statistic: MonthsStat = {}

pages: Set[str] = set()
genders: List[str] = ["Male", "Female"]
devices: List[str] = ["Mobile", "Desktop"]
    
# name.csv ниже нужно заменить на название csv файла, который вы хотите открыть
with open(file_path, mode='r') as csv_file:
    csv_reader = csv.DictReader(csv_file, fieldnames=['ID','page','date','device','gender'])
    for row in csv_reader:
        _, page, date, device, gender = row.values()
        year, month, day = date.split("-")
        assert device in devices, f"Unexpected device '{device}'"
        assert gender in genders, f"Unexpected gender '{gender}'"
        pages.add(page)
        
        device_stat = statistic.setdefault(month, {})
        gender_stat = device_stat.setdefault(device, {})
        pages_stat = gender_stat.setdefault(gender, {})
        stat = pages_stat.setdefault(page, Stat())
        stat.actions += 1

# Normalize:
if normalize:
    for device_stat in statistic.values():
        for gender_stat in device_stat.values():
            for pages_stat in gender_stat.values():
                total_actions = max([value.actions for value in pages_stat.values()])
                for stat in pages_stat.values():
                    stat.actions = round(100 * stat.actions / total_actions, 2)
        
# Printing statistic:
print("Month\tDevice\tPage\t\tMale\tFemale\tDelta")
for month in sorted(statistic.keys()):
    print(f"Month {month}:")
    for device in devices:
        print(f"\tOn {device}:")
        gender_stat = statistic[month][device]
        for page in sorted(pages):
            male = gender_stat["Male"][page].actions
            female = gender_stat["Female"][page].actions
            delta = abs(male - female)
            print(f"\t\t{page}:\t{male}\t{female}\t{round(delta, 2)}")
```

Судя по всему, в данном случае пол посетителя не имеет значения, т.к. на каждом этапе разница в конверсии между обоими полами не превышает 2% (см. крайний правый столбец "Delta")
